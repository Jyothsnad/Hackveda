```python
# Generate Random Dataset

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
x = 2 - 3 * np.random.normal(0,1,20)
y = x - 2 * (x ** 2) + 0.5 * (x ** 3) + np.random.normal(-3,3,20)
plt.scatter(x,y)
plt.show()
```

```python
# Apply Linear Regression

from sklearn.linear_model import LinearRegression
lr = LinearRegression()
from sklearn.metrics import r2_score

x = x[:,np.newaxis]
y = y[:,np.newaxis]

lr.fit(x,y)
ypred = lr.predict(x)

plt.scatter(x,y)
plt.plot(x,ypred,color='r')
plt.show()

print('R Squared : ',r2_score(y,ypred))
```

```python
import operator
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import PolynomialFeatures

polynomial_features = PolynomialFeatures(degree=2)
xpoly = polynomial_features.fit_transform(x)

lr.fit(xpoly,y)
ypoly_pred = lr.predict(xpoly)

mse = np.sqrt(mean_squared_error(y,ypoly_pred))
r2 = r2_score(y,ypoly_pred)
print(mse)
print(r2)

plt.scatter(x,y)
# sort the values of x before line plot
sort_axis = operator.itemgetter(0)
sorted_zip = sorted(zip(x,ypoly_pred), key=sort_axis)
x, ypoly_pred = zip(*sorted_zip)
plt.plot(x, ypoly_pred, color='m')
plt.show()
```
