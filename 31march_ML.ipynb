```python
# Convert Binary text category to numbers using Label Binarizer

from sklearn.preprocessing import LabelBinarizer

# Create an object for LabelBinarizer
lb = LabelBinarizer()
df.Status = lb.fit_transform(df.Status)
df
```

```python
# https://en.wikipedia.org/wiki/One-hot  ---> for n-categorical data into numbers

print(df.Result)
df_new = pd.get_dummies(df.Result, prefix='Result')
```

```python
df_new
```

```python
# Concatenate df with df_new

df = pd.concat([df,df_new], axis=1)
df
```

```python
# Delete Result text category from df

del(df['Result'])
df
```
